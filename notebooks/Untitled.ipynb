```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Determine the absolute path to the data/processed directory

project_dir = '/Users/alperen.us/customer_experience_project/'
data_processed_path = os.path.join(project_dir, 'data/processed')

# Load the feature-engineered customer data
df_customers = pd.read_csv(os.path.join(data_processed_path, 'feature_engineered_customer_data.csv'))
```

```python
# Display basic statistics
df_customers.describe()
```

```python
# Check for any remaining missing values
df_customers.isnull().sum()
```

```python
# Plot the distribution of Age Groups
plt.figure(figsize=(8, 6))
sns.countplot(x='AgeGroup', data=df_customers)
plt.title('Distribution of Age Groups')
plt.show()
```

```python
# Reorder Income Bracket categories logically
order = ['Low', 'Lower-Middle', 'Middle', 'Upper-Middle', 'High']

# Plot the relationship between Income Bracket and Spending Score
plt.figure(figsize=(8, 6))
sns.boxplot(x='IncomeBracket', y='SpendingScore', data=df_customers, order=order)
plt.title('Income Bracket vs. Spending Score')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df_customers.select_dtypes(include=[float, int])

# Display the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
